## Data Source and Contents

```python
# import necessary libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
# load dataset
movies_data = pd.read_csv('https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv')
movies_data
```

## Building Recommender System

```python
# convert dataset to dictionary with titles as keys
movies_dict = movies_data.set_index('Movie Name').to_dict(orient='index')

# extract movie titles from dictionary
movies_titles = list(movies_dict.keys())
```

```python
# select columns to train model
train_columns = ['IMDB Rating', 
                   'Biography',
                   'Drama', 
                   'Thriller', 
                   'Comedy',
                   'Crime',
                   'Mystery', 
                   'History']

x_train = movies_data[train_columns]
x_train
```

```python
# build nearest neighbors model
neigh = NearestNeighbors(n_neighbors=5)
model = neigh.fit(x_train)
```

## Test the model

```python
# create dataframe for test movie "The Post"
the_post = pd.DataFrame({'IMDB Rating':[7.2], 
                         'Biography':[1], 
                         'Drama':[1], 
                         'Thriller':[0], 
                         'Comedy':[0], 
                         'Crime':[0], 
                         'Mystery':[0], 
                         'History':[1]})

# first array contains distances of the recommended movie, second array contains index of the recommended movie
rec_distances, rec_indices = model.kneighbors(the_post)
```

## Print the Results

```python
# print results
print("More Movies Like \"The Post\":")

# loop through indexes and print the associated movie title
for i, idx in enumerate(rec_indices[0]):
    print(f"{movies_titles[idx]} (Distance = {rec_distances[0][i]})")
```
